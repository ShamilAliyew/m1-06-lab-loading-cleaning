```python
import pandas as pd
```

```python
tickets = pd.read_csv('data_safe_copy.csv')
tickets.shape
```

```python
tickets.head()
```

```python
tickets.info()
```

```python
tickets['opened_at'] = pd.to_datetime(tickets['opened_at'], errors='coerce')
tickets[['category', 'priority']] = (tickets[['category', 'priority']].apply(lambda col: col.str.strip().str.lower()))
```

```python
print(tickets['category'].unique())
print(tickets['priority'].unique())
```

```python
tickets["resolution_minutes"] = pd.to_numeric(tickets["resolution_minutes"],errors="coerce")
tickets.head()
```

```python
missing_count = tickets['resolution_minutes'].isna().sum()
print(missing_count)
```

```python
tickets_clean = tickets.dropna(subset=['resolution_minutes'])

print("Before cleaning:", len(tickets))
print("After cleaning:", len(tickets_clean))
```

```python
(tickets['resolution_minutes']>=0).all()
```

```python
print("Categories:", tickets_clean["category"].unique())
print("Priorities:", tickets_clean["priority"].unique())
```

```python
total_records = len(tickets)
clean_records = len(tickets_clean)
missing_resolution = tickets["resolution_minutes"].isna().sum()
avg_resolution = tickets_clean["resolution_minutes"].mean()
max_resolution = tickets_clean["resolution_minutes"].max()


summary = {
    "total_records": total_records,
    "clean_records": clean_records,
    "missing_resolution": missing_resolution,
    "avg_resolution": avg_resolution,
    "max_resolution": max_resolution
}

for k,v in summary.items():
    print(k,':',v)
```

```python
print(
    "Check:",
    clean_records + missing_resolution == total_records
)
```
